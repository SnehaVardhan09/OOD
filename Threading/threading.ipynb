* Threading whenever its going to significantly speed-up our program
* Speed-up comes from running different tasks 

```python
import time 

start = time.perf_counter()

def do_something():
    print('Sleeping 1 sec...')
    time.sleep(1)
    print('Done Sleeping...')

do_something()
do_something()


finish = time.perf_counter()

print(f'Finished in {round(finish - start, 2)} sec')
```

* Running Synchronously so the total time taken is 2 sec
* 
* CPU bound(crunching a lot of numbers and using cpu) and i/o bound(not using cpu. reading writing from the file system)
* When it comes to *threading* we're gonna see benefits when our tasks are IO bound 
* We are not getting much of benefit from threading if we are using CPU bound operations 
* When it is CPU bound we use Multi-processing and run it in parallel
* 


```python
import threading
import time 

start = time.perf_counter()

def do_something():
    print('Sleeping 1 sec...')
    time.sleep(1)
    print('Done Sleeping...')


t1 = threading.Thread(target = do_something)
t2 = threading.Thread(target = do_something)

t1.start()
t2.start()

t1.join()
t2.join()

finish = time.perf_counter()

print(f'Finished in {round(finish - start, 2)} sec')
```

```python
## running 10 functions - synchronously takes 10sec but threading takes 1 sec

import threading
import time 

start = time.perf_counter()

def do_something():
    print('Sleeping 1 sec...')
    time.sleep(1)
    print('Done Sleeping...')

threads =[]
for _ in range(10):
    t = threading.Thread(target = do_something)
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()

finish = time.perf_counter()
print(f'Finished in {round(finish - start, 2)} sec')
```
