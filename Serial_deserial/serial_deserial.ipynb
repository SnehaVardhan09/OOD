* Serialization - Process of converting python data types to JSON format.
* Deserialization - Process of converting JSON to python data types.


```python
import json 

L = [1, 2,3 ,4]

with open('Demo.json', 'w') as f:
    json.dump(L,f)
```

```python
d = {'name' : 'Sneha', 'age': 27, 'gender': 'male'}

with open('demo.json', 'w') as f:
    json.dump(d, f, indent=4)
```

```python
#Deserialization
with open('demo.json', 'r') as f:
    print(json.load(f))
```

### Serializing and Deserializing custom objects

```python
class Person:

    def __init__(self, fname, lname, age, gender ):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.gender = gender

person = Person('sneha', 'vardhan', 27, 'male')
```

```python
with open('demo.json' , 'w') as f:
    json.dump(person, f)
```

```python
def show_object(person):
    if isinstance(person, Person):
        #return "{} {} age -> {} gender -> {}".format(person.fname, person.lname, person.age, person.gender)
        return {'name':person.fname + ''+ person.lname, 'age':person.age, 'gender':person.gender}

with open('demo.json' , 'w') as f:
    json.dump(person, f, default=show_object)
```

```python
with open('Demo.json', 'r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

### Pickling
* `Pickling` is the process where by a python object hierarchy ---> byte stream
*  `Unpickling` byte stream(from a binary file or bytes-like object) ---> object hierarchy.

```python
class Person:

    def __init__(self, name, age, gender ):
        self.name = name
        self.age = age
        self.gender = gender

    def DisplayName(self):
        print('Hi my name is ', person.name,' age', person.age, 'gender', person.gender)

person = Person('sneha', 27, 'male')
```

```python
import pickle 

with open('person.pkl','wb') as f:
    pickle.dump(person,f)
```

```python
with open('person.pkl', 'rb') as f:
    p = pickle.load(f)

p.DisplayName()
```

* Pickle lets the user to store data in binary format. Json lets the user store data in a human-readable text format
