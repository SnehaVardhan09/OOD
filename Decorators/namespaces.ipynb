### Namespaces
* Is space that holds names(identifiers). Programmatically speaking, namespaces are dictionary of identifiers (keys) and their objects(values)
* 4 types: Builtin / Global / Enclosing / Local (LEGB)
#### Scope & LEGB Rule
* A scope is a textual region of a python program where a namespace is directly accessible.

* The interpreter searches for a name from the inside out, looking in the local, enclosing, global and finally the built-in scope. if the interpreter doesnt find the name in any of these locations, then Python raises a NameError exception.


```python
# local and global

a =2 #global

def temp():
    b =3  #local
    print(b) 

temp()
print(a)
```

```python
# local and global

a =2 #global

def temp():
    a =3  #local
    #print(b) 

temp()
print(a)
```

```python
# local and global -> local doesnt have but global has

a =2 #global

def temp():
   
    print(a)     # local can access global

temp()
print(a)
```

```python
# local and global -> editting global

a =2 #global

def temp():
   #local var
   a+=1
   print(a)   # local can read global but cannot assign

temp()
print(a)
```

```python
a =2 #global

def temp():
   #local var
   global a
   a+=1
   print(a)   # local can read global but cannot assign but it can change after "global"

temp()
print(a)
```

```python
def temp():
   #local var
   global a
   a = 1   
   print(a)   

temp()
print(a)
```

```python
def temp(z):
   #local var
   print(z)   

a =5
temp(5)
print(a)
```

### Decorators
* A decorator in python is a function that receives another function as input and adds some functionality to it and returns it
* This can happen only because python function are 1st class citizens.
* There are 2 types of decorators avaliable in python.
     * Builtin decorators like @staticmethod @classmethod, @abstractmethod, @property etc
     * User defined decorators that we programmers can create according to our needs

```python
# python are first class citizens

def modify(func, num):
    return func(num)

def square(num):
    return num**2

modify(square, 2)
```

```python
def my_decorator(func):
    def wrapper():
        print("***************************")
        func()
        print("***************************")
    return wrapper

def hello():
    print('hello')

a = my_decorator(hello)
a()
```
